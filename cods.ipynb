```python
import numpy as np # Lineer cebir kütüphanesi
import pandas as pd # Veri işleme ve CSV dosya okuma kütüphanesi
```

```python
# data setini yükleme
df = pd.read_json("News_Category_Dataset_v3.json",lines=True)
df
```

```python
# Veri kümesindeki boş hücreleri kontrol etme
df.isna().sum()
```

```python
# Her kategori için kayıt sayısını kontrol etme
df['category'].value_counts()
```

```python
# Veri kümesi büyük olduğundan, 5 kategori seçiyorum ve her kategori için 5000 kayıt seçiyorum
columns_lst = ['POLITICS','WELLNESS','ENTERTAINMENT','TRAVEL','STYLE & BEAUTY']
```

```python
# columns_list'te listelenen kategorilere ait kayıtları seçme
df_t = df[df['category'].isin(columns_lst)]
df_t
```

```python
# Her kategori için 5000 kayıt seçme
df_sample = df_t.groupby('category').head(5000)
df_sample
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Kategorileri sayısal değerlere dönüştürme
df_sample_copy = df_sample.copy()
df_sample_copy['target'] = le.fit_transform(df_sample_copy['category'])
df_sample_copy.head()
```

```python
# Sınıflar için sözlük oluşturma
classes_dict = {le.classes_[i] : i for i in range(len(le.classes_))}
classes_dict
```

```python
# Başlık ve kısa açıklama sütunlarını yeni bir 'text' sütununda birleştirme
df_sample_copy['text'] = df_sample_copy['headline'] +" " + df_sample_copy['short_description']
df_sample_copy['text'].iloc[20]
```

```python
from sklearn.model_selection import train_test_split

# Veri kümesini eğitim ve test setlerine bölmek
X_train, X_test, y_train, y_test = train_test_split(df_sample_copy['text'], df_sample_copy['target'], test_size=0.2,random_state=26)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import spacy
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# SpaCy'nin İngilizce dil modelini yükleme
nlp = spacy.load("en_core_web_sm")
```

```python
# Stop words ve noktalama işaretlerini filtreleyen ve lemmaları çıkaran bir işleme fonksiyonu tanımlama
def preprocesser(text_array):
    preprocessed_texts = []
    for text in text_array:
        doc = nlp(text)
        words_lst = []
        for token in doc:
            if not token.is_stop and not token.is_punct:
                words_lst.append(token.lemma_)
        preprocessed_text = " ".join(words_lst)
        preprocessed_texts.append(preprocessed_text)
    return preprocessed_texts
```

```python
# FunctionTransformer sınıfı, bir işlevi bir dönüştürücü olarak kullanmanıza izin veren bir sınıftır.
# Bu durumda, preprocesser adlı fonksiyon, metin verilerini önceden işlemek üzere bir dönüştürücü olarak kullanılıyor.
# preprocesser fonksiyonu, her bir metni alır, SpaCy dil modelini kullanarak stop kelimeleri ve noktalama işaretlerini filtreler ve lemmaları çıkarır
text_processer = FunctionTransformer(preprocesser)
```

```python
cls1 = make_pipeline(
    text_processer,
    CountVectorizer(),
    MultinomialNB()
)



cls1.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls1.predict(X_test)))
```

```python
cls2 = make_pipeline(
    text_processer,
    CountVectorizer(ngram_range=(1,3)),
    MultinomialNB()
)



cls2.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls2.predict(X_test)))
```

```python
cls3 = make_pipeline(
    text_processer,
    CountVectorizer(),
    LogisticRegression(multi_class='multinomial', solver='lbfgs')
)



cls3.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls3.predict(X_test)))
```

```python
cls4 = make_pipeline(
    text_processer,
    CountVectorizer(ngram_range=(1,3)),
    LogisticRegression(multi_class='multinomial', solver='lbfgs')
)



cls4.fit(X_train, y_train)
```

```python
print(classification_report(y_test,cls4.predict(X_test)))
```
